```python
import re

import sklearn

from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer

from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.model_selection import train_test_split

from sklearn.metrics.pairwise import cosine_similarity

from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
stop_words = stopwords.words('english')
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()
vectorizer = CountVectorizer(analyzer='word', ngram_range=(1, 1))
tfidf_transformer = TfidfTransformer()
```

```python
title = df_vinhos.title
descricao  = df_vinhos.description.apply(lambda x : [n.strip() for n in x.split(',')]).tolist()
```

```python
descricao
```

```python
title
```

```python
# função para tratar cada palavra e remover stopwords
def processamento(tokens):
    
    token_processado = []
    for token in tokens:
        token = token.lower()
        token = lemmatizer.lemmatize(token)
        
        if token not in stop_words:
            token = stemmer.stem(token)
            token_processado.append(token)
        
    return token_processado

documentos = []
for titulos in title:
    
    # expressão regular para remover pontuações do texto
    descricao = re.sub(r'[^\w\s]','', titulos)
    tokens = processamento(word_tokenize(titulos))
    
    documentos.append(' '.join(tokens))
```

```python
Count_titulos = vectorizer.fit_transform((df_vinhos['title'][0:20000]))
cos_titulos = cosine_similarity(Count_titulos)
print(cos_titulos)
```

```python
def find_title_from_index(index): 
    return df_vinhos[df_vinhos.index == index]["title"].values[0] 
def find_index_from_title(title): 
    return df_vinhos[df_vinhos.title == title].index.values[0]
```

```python
u = input(str('Digite o titulo do vinho: ')) # exemplo > Bodega El Transito 2012 Pietro Marini Torrontés (Salta)

wine_index = find_index_from_title(u)
similar_wine = list(enumerate(cos_titulos[wine_index]))
sorted_similar_wine = sorted(similar_wine, key=lambda x : x [1],reverse=True)[1:]
i=0 
lista = []
for elemento in sorted_similar_wine: 
    i=i+1 
    lista.append(find_title_from_index(elemento[0]))

    if i>5: 
        break

# criando um dataframe com os vinhos similares
df_similares = pd.DataFrame()
for k in range(len(lista)):
    df_similares = pd.concat([df_similares, df_vinhos[df_vinhos['title'] == lista[k]]])

# removendo todos os valores repetidos
df_similares = df_similares.drop_duplicates (subset = None, keep = 'first', inplace = False)
df_similares
```

```python
ordem_precos = df_similares.sort_values(by=['price'],ascending=False)

gf_20 = px.bar(ordem_precos, x='title', y='price', color='title', text_auto = True, 
               title='<span style="font-size: 19.5px;"><b>Vinhos semelhantes ao Urban Uco Malbec</b>' + '<br>' +  
                     '<span style="font-size: 14px;"> Os vinhos estão ordenados do mais caro ao mais barato</span>',
               labels={'title':'<b>Vinho</b>', 'price':'<b>Preço em Dolares</b>'},
               color_discrete_sequence= px.colors.sequential.Teal_r,
               height=350
)

gf_20.update_layout()
gf_20.update_layout(xaxis={'visible':False})
o = 0
```
